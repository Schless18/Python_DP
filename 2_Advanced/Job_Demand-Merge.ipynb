```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
```

```python
df_US['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')
df_US
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot
```

```python
df_US_pivot.reset_index(inplace=True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month

df_US_pivot.sort_values('month_no', inplace=True)
df_US_pivot.set_index('job_posted_month', inplace=True)
df_US_pivot.drop(columns='month_no', inplace=True)
df_US_pivot
```

```python
top_3 = df_US['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()

top_3 = df_US_pivot.sum().sort_values(ascending=False).head(3).index.tolist()
top_3
```

```python
df_US_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```

```python
df_US_software_pivot = pd.read_csv('https://lukeb.co/software_csv', index_col='job_posted_month')
df_US_software_pivot
```

```python
df_US_merged = df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
df_US_merged
```

```python
top_5 = df_US_merged.sum().sort_values(ascending=False).head(5).index.to_list()
top_5
```

```python
df_US_merged[top_5].plot(kind='line')
plt.title('Top five Job Postings in the US by count')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 30000)
plt.legend()
plt.show()
```
